```python
# Matching the similar products from the Flipkart dataset with the Amazon dataset with Merge function
Mrg=pd.merge(Amz,Flip,on=["product_name"],how='inner')
Mrg
```

```python
result=Mrg[['product_name','retail_price_x','discounted_price_x','product_name','retail_price_y','discounted_price_y']]
result
```
